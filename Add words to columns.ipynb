```python
import pandas as pd
import pymorphy2

# Прочитаем CSV файл
file_path = 'merged_frequencies - словник для объединения.csv'
data = pd.read_csv(file_path)

# Определим части речи для каждого слова в первом столбце
morph = pymorphy2.MorphAnalyzer()
first_column_words = data['Unnamed: 0'].dropna().tolist()
parts_of_speech = [(word, morph.parse(word)[0].tag.POS) for word in first_column_words]

# Создадим словари для разных частей речи
nouns = [word for word, pos in parts_of_speech if pos == 'NOUN']
verbs = [word for word, pos in parts_of_speech if pos in ('VERB', 'INFN')]
adjectives = [word for word, pos in parts_of_speech if pos in ('ADJF', 'ADJS')]
adverbs = [word for word, pos in parts_of_speech if pos == 'ADVB']

# Добавим слова в соответствующие столбцы и уберем дубликаты
existing_nouns = data['Существительные'].dropna().tolist()
existing_verbs = data['Глаголы'].dropna().tolist()
existing_adjectives = data['Прилагательные'].dropna().tolist()
existing_adverbs = data['Наречия'].dropna().tolist()

# Объединяем списки и убираем дубликаты
new_nouns = list(set(existing_nouns + nouns))
new_verbs = list(set(existing_verbs + verbs))
new_adjectives = list(set(existing_adjectives + adjectives))
new_adverbs = list(set(existing_adverbs + adverbs))

# Создаем новый DataFrame
new_data = pd.DataFrame({
    'Существительные': pd.Series(new_nouns),
    'Глаголы': pd.Series(new_verbs),
    'Прилагательные': pd.Series(new_adjectives),
    'Наречия': pd.Series(new_adverbs)
})

# Сохраним результат в новый CSV файл
output_file_path = 'updated_frequencies.csv'
new_data.to_csv(output_file_path, index=False)

output_file_path
```
